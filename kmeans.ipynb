```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset
from sklearn.cluster import KMeans
```

```python
df=dataset
```

```python
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
```

```python
df=df.drop('CustomerID',axis=1)
```

```python
df
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Scaling the data to normalize
model = KMeans(n_clusters=2).fit(df)
```

```python
plt.scatter(df.iloc[:, 1],df.iloc[:, 2], c=model.labels_.astype(float))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.array(df['Annual Income (k$)'])
y = np.array(df['Genre'])
z = np.array(df['Spending Score (1-100)'])

ax.scatter(x,y,z, marker="X", c=model.labels_.astype(float))

plt.show()
```
