```python
x=data.iloc[:,0:2]
```

```python
x=x.values
```

```python
x
```

```python

tr={}
for i in range(len(x)):
    if  x[i][0]  in tr.keys():
        tr[x[i][0]].append(x[i][1])
    else :
        tr[x[i][0]]=[x[i][1]]
```

```python
dset=[]
```

```python
for key,value in tr.items():
    dset.append(value)
```

```python
te = TransactionEncoder()
te_ary = te.fit(dset).transform(dset)
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

```python
frequent_itemsets = apriori(df, min_support=0.005, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="confidence", min_threshold=0.005)
```
